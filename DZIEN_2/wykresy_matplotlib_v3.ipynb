```python
import numpy as np
import matplotlib.pyplot as plt
```

# wykres słupkowy

```python
fig,ax = plt.subplots()
marki = ['audi','toyota','mercedes','honda','kia','jeep','seat']
sprzedaz = [12.7,28.9,9.8,16.8,23.5,7.8,28.3]

bar_labels = ['audi','toyota','mercedes','honda','kia','jeep','seat']
bar_colors = ['tab:red','tab:green','tab:red','tab:blue','tab:orange','#ff9966','tab:green']

ax.bar(marki,sprzedaz,label=bar_labels,color=bar_colors)
ax.set_ylabel('Sprzedaż wybranych marek')
ax.set_xlabel('marki samochodów')

ax.legend(title = 'marki')
plt.show()
```

# wykres kołowy

```python
labels = 'Warszawa','Kraków','Lublin','Katowice','Gdańsk'
sizes = [56,34,26,24,30]
explode = (0,0.1,0,0,0.2)
```

```python
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax.axis('equal')
plt.show()
```

# wykresy funkcji falowych nałożone na siebie

```python
#deklaracja danych
x1 = np.linspace(0.0,5)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)

x2 = np.linspace(0.0,2)
y2 = np.cos(2*np.pi*x2)
```

```python
fig,(ax1,ax2) = plt.subplots(2,1)
fig.suptitle('dwa wykresy: tłumienie i brak')

ax1.plot(x1,y1,'o-')
ax1.set_ylabel('amp drgań tłumionych')

ax2.plot(x2,y2,'.-')
ax2.set_ylabel('amp drgań nietłumionych')
ax2.set_xlabel('czas [s]')

plt.show()
```
