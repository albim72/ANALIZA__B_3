```python
import numpy as np
import matplotlib.pyplot as plt
```

# wykres słupkowy

```python
fig,ax = plt.subplots()
marki = ['audi','toyota','mercedes','honda','kia','jeep','seat']
sprzedaz = [12.7,28.9,9.8,16.8,23.5,7.8,28.3]

bar_labels = ['audi','toyota','mercedes','honda','kia','jeep','seat']
bar_colors = ['tab:red','tab:green','tab:red','tab:blue','tab:orange','#ff9966','tab:green']

ax.bar(marki,sprzedaz,label=bar_labels,color=bar_colors)
ax.set_ylabel('Sprzedaż wybranych marek')
ax.set_xlabel('marki samochodów')

ax.legend(title = 'marki')
plt.show()
```

# wykres kołowy

```python
labels = 'Warszawa','Kraków','Lublin','Katowice','Gdańsk'
sizes = [56,34,26,24,30]
explode = (0,0.1,0,0,0.2)
```

```python
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax.axis('equal')
plt.show()
```
