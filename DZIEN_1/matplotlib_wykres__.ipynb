```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0,2.0,0.01)
s = 1 + np.sin(2*np.pi*t)
```

```python
fig,ax = plt.subplots()
ax.plot(t,s)
ax.set(xlabel='czas [s]',ylabel='napięcie prądu [mV]', title='pomoar napięcia prądu w czasie t')
ax.grid()
fig.savefig('wykres.png')
plt.show()
```
